How would you collect data on a stock's performance without a computer doing it automatically? Go to [this website](https://www.tradingview.com/) and look up a stock (any stock). You are going to write a function that takes live input of the stock's value (entered by you), saves all inputs in a list, stops once you enter 10 values **or** three consecutive decreasing values (whichever comes first), and returns the values.

For this problem:

- Call your function `my_stock`
- Make sure that you use the `input()` option within your function to accept entries
- Make sure you are saving all entries in a list
- Use `if-else` and a `while` loop

**Note**: There may be multiple ways to get this function to work, but to receive full points, your function **must** have the five components listed above (there may be more efficient ways to do it, but I want you to show me you know how to use *these methods*).

```python
def my_stock():
    """
    Prompts the user to enter stock values until either 10 values are entered 
    or 3 consecutive decreasing values are entered. 

    The function keeps track of consecutive decreasing stock values and stops 
    collecting data if the condition is met before reaching 10 entries.

    Returns:
        list: A list of stock values entered by the user.
    """
    stock_values = []  # List to store the entered stock values
    consecutive_decreasing = 0  # Counter for consecutive decreasing values
    
    # Continue to collect stock values until either 10 values are entered
    # or there are 3 consecutive decreasing values
    while len(stock_values) < 10 and consecutive_decreasing < 3:   
        stock_values.append(float(input("Enter the stock value: ")))  
        
        # Check if the current value is less than the previous one
        if len(stock_values) > 1:
            if stock_values[-1] < stock_values[-2]: 
                consecutive_decreasing += 1  
            else:
                consecutive_decreasing = 0  
        
    return stock_values 

# Call the function and store the result
manual_data_entry = my_stock()

# Display the entered stock values
manual_data_entry
```

```python
# getting the whole row and calculating the BMW
def calculate_BMI(pokemon):
    height = pokemon.get("Height",0)
    weight = pokemon.get("Weight",0)
    height_modified = height * height
    return 703 * (weight/height_modified)

# getting the whole row and seeing if has a baseset of 70 >= or is not a firetype
def has_baseSet(pokemon):
    if pokemon.get("BaseSpeed", 0) >= 70 and pokemon.get("MainType", "") != "Fire":
        return True
    return False

# Part 1 adding a new column
for poke in df_dict:
    poke["BMI"] = calculate_BMI(poke)

print(df_dict)

# Part 2 excluding the base sets
new_pokemon = []
for poke in df_dict:
    if has_baseSet(poke):
        new_pokemon.append(poke)

print(new_pokemon)
```
