```python
df_fifa.Nationality.value_counts()
```

```python
df_filtered = df_fifa[df_fifa['Nationality'].isin(['England', 'Germany', 'Spain'])]

df_grouped = df_filtered.groupby(['Nationality', 'Age'])['Overall'].mean().reset_index()

fig = px.scatter(df_grouped, x='Age', y='Overall', color='Nationality',
                 labels={'Overall': 'Mean Overall Rating', 'Age': 'Player Age'})

```

```python
import numpy as np

a = np.array([2, -1, 3])
b = np.array([-4, -2, 0])
c = np.array([3, 3, -3])
```

```python
b_plus_c = b + c
magnitude = np.linalg.norm(b_plus_c)
print("||b + c|| =", magnitude)

times_add = 2 * a + b
print("2a + b =", times_add)

dot= np.dot(c, a)
print("c x a =", dot)

a_div_2_minus_c = (a / 2) - c
magnitude = np.linalg.norm(a_div_2_minus_c)
print("||a/2 - c|| =", magnitude)

# 5. Compute b · (a + c)
dot = np.dot(b, a + c)
print("b · (a + c) =", dot)
```
